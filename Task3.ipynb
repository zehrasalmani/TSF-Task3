```python
import pandas as pd
from sklearn import datasets
```

```python
#loading the dataset
iris = datasets.load_iris()
```

```python
#making iris dataframe with indedpendent values
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
# target values i.e the dependent variable are stored in y variable
y=iris.target
y
```

```python
# Defining Decision Tree Algorithm
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()                           #decision tree object
train = dt.fit(df,y)                                  #training the data       
```

# visualizing the Decision Tree

```python
# Import necessary libraries for graph viz

from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
```

```python
#visualizing the graph
from six import StringIO
dot_data = StringIO()
export_graphviz(train, out_file=dot_data, feature_names=iris.feature_names,special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```
